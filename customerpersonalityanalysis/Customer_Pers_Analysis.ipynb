```python
#deal with outliers using the IQR method for the income column
q1=df['Income'].quantile(0.25)

q3=df['Income'].quantile(0.75)

iqr=q3-q1

q1,q3,iqr

upper_limit=q3+(1.5*iqr)

lower_limit=q1-(1.5*iqr)

lower_limit, upper_limit
df.loc[(df['Income']>upper_limit)|(df['Income']<lower_limit)]
df.loc[df['Income']>upper_limit,'Income']=upper_limit
df.loc[df['Income']<lower_limit,'Income']=lower_limit
```
